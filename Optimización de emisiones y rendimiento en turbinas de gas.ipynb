# Proyecto - Optimización de Emisiones y Rendimiento en Turbinas de Gas

Una **turbina de gas** es una máquina térmica que convierte la energía química de un combustible en energía mecánica y, posteriormente, en energía eléctrica. Su funcionamiento se basa en la compresión de aire, la mezcla con combustible y la combustión a alta temperatura, generando gases que expanden a través de una turbina y producen trabajo útil.

Estas turbinas se utilizan principalmente para:

- **Generación de electricidad**
- **Producción de energía en centrales térmicas**
- **Sistemas de cogeneración**
- **Procesos industriales con alta demanda energética**
- **Aplicaciones en petróleo, gas, petroquímica y utilities**

Desde el punto de vista operativo, las turbinas de gas son equipos de gran relevancia porque su desempeño depende de variables ambientales, térmicas y de presión que afectan tanto la **eficiencia energética** como el **nivel de emisiones contaminantes**.

En este proyecto se analizará un conjunto de datos asociado a una turbina de gas con el objetivo de estudiar las variables que influyen sobre su rendimiento y sobre las emisiones de **monóxido de carbono (CO)** y **óxidos de nitrógeno (NOx)**. El análisis busca identificar patrones, relaciones entre variables y oportunidades para apoyar decisiones de operación y optimización en entornos energéticos e industriales.

## Descripción de las variables del dataset

A continuación se presentan las variables incluidas en el dataset, junto con su significado, unidad de medida y relevancia operativa dentro del funcionamiento de una turbina de gas.

| Sigla | Variable | Unidad | Importancia en la turbina |
|---|---|---|---|
| **AT** | Temperatura ambiente | °C | Afecta la densidad del aire de admisión y, por tanto, el rendimiento del compresor y la potencia disponible de la turbina. |
| **AP** | Presión ambiente | mbar | Influye en la masa de aire que ingresa al sistema y en la eficiencia global del ciclo de la turbina. |
| **AH** | Humedad ambiente | % | Modifica las propiedades del aire de entrada y puede influir en el proceso de combustión y en el nivel de emisiones. |
| **AFDP** | Diferencia de presión en el filtro de aire | mbar | Puede reflejar obstrucción o ensuciamiento en el sistema de admisión, afectando el caudal de aire y el desempeño operativo. |
| **GTEP** | Presión de escape de la turbina de gas | mbar | Permite caracterizar las condiciones de descarga del sistema y puede relacionarse con el comportamiento del flujo de salida. |
| **TIT** | Temperatura de entrada a la turbina | °C | Es una variable crítica para el rendimiento térmico, la potencia generada y la formación de emisiones contaminantes. |
| **TAT** | Temperatura después de la turbina | °C | Ayuda a evaluar el comportamiento térmico del proceso y la energía extraída durante la expansión de los gases. |
| **TEY** | Energía generada por la turbina | MWh | Representa el rendimiento energético del equipo y funciona como indicador de desempeño operativo. |
| **CDP** | Presión de descarga del compresor | mbar | Refleja el nivel de compresión del aire antes de la combustión y es clave en la operación del ciclo Brayton. |
| **CO** | Monóxido de carbono | mg/m³ | Emisión asociada a combustión incompleta. Es relevante para evaluar eficiencia de combustión e impacto ambiental. |
| **NOx** | Óxidos de nitrógeno | mg/m³ | Emisiones altamente relevantes en términos regulatorios y ambientales, estrechamente ligadas a la temperatura de combustión. |


## Interpretación general de las variables

Las variables del dataset pueden agruparse en tres grandes bloques analíticos:

### 1. Variables ambientales
Incluyen: **AT**, **AP**, **AH** y **AFDP**.  
Estas variables describen las condiciones de entrada del sistema y del aire de admisión. Su análisis es importante porque afectan la densidad del aire, la calidad del flujo de entrada y, por ende, el rendimiento general de la turbina.

### 2. Variables operativas y térmicas
Incluyen: **GTEP**, **TIT**, **TAT**, **TEY** y **CDP**.  
Estas variables reflejan el comportamiento interno de la turbina en términos de compresión, combustión, expansión y generación energética. Son fundamentales para evaluar desempeño, eficiencia térmica y estabilidad de operación.

### 3. Variables de emisiones
Incluyen **CO** y **NOx**.  
Estas variables representan los principales indicadores de impacto ambiental del sistema. Es decir, su análisis es clave para evaluar la calidad de la combustión, el cumplimiento de las leyes ambientales y el equilibrio entre desempeño energético y sostenibilidad operativa.

## 1.3 Comparación de las características del dataset filtrado con respecto a la producción de Monóxido de carbono (CO), Oxidos de Nitrógeno (NOX) y Energía Generada por la Turbina (TEY) usando Gráficos de Dispersión (Scatter Plots)

```python
# Separo en listas las columnas objetivo y las características
features = ['AT', 'AP', 'AH', 'AFDP', 'GTEP', 'TIT', 'CDP']
targets = ['CO', 'NOX', 'TEY']

# Agrego en diccionarios los nombres completos de las variables y sus unidades para nombrar mejor los gráficos
feature_names = {
    'AT': 'Temperatura ambiente',
    'AP': 'Presión ambiente',
    'AH': 'Humedad ambiente',
    'AFDP': 'Dif. presión filtro de aire',
    'GTEP': 'Presión de escape',
    'TIT': 'Temperatura entrada turbina',
    'CDP': 'Presión descarga compresor'
}

feature_units = {
    'AT': '°C',
    'AP': 'mbar',
    'AH': '%',
    'AFDP': 'mbar',
    'GTEP': 'mbar',
    'TIT': '°C',
    'CDP': 'mbar'
}

target_names = {
    'CO': 'Monóxido de carbono',
    'NOX': 'Óxidos de nitrógeno',
    'TEY': 'Energía generada'
}

target_units = {
    'CO': 'mg/m³',
    'NOX': 'mg/m³',
    'TEY': 'MWh'
}

# Creo la figura según el tamaño de las características y las variables objetivo.

fig, axes = plt.subplots(len(features), len(targets), figsize=(18, 38))

# Recorro cada característica junto con su posición en la lista
for i, feature in enumerate(features):
    
    # Recorro cada variable objetivo junto con su posición en la lista
    for j, target in enumerate(targets):
        
        # Selecciono el subplot correspondiente según la fila y la columna
        ax = axes[i, j]
        
        # Creo un gráfico de dispersión para observar la relación entre ambas variables
        ax.scatter(df_high_tey[feature], df_high_tey[target], alpha=0.25, color='salmon')
        
        # Asigno la etiqueta del eje X con la sigla y la unidad de la característica
        ax.set_xlabel(f"{feature} ({feature_units[feature]})")
        
        # Asigno la etiqueta del eje Y con la sigla y la unidad de la variable objetivo
        ax.set_ylabel(f"{target} ({target_units[target]})")
        
        # Defino el título del subplot usando el nombre completo y la sigla de ambas variables
        ax.set_title(
            f"{feature_names[feature]} ({feature}) vs {target_names[target]} ({target})",
            fontsize=13
        )
        
        # Calculo el coeficiente de correlación de Pearson entre ambas variables
        corr = df0[feature].corr(df0[target])
        
        # Agrego debajo del gráfico el valor de la correlación calculada
        ax.text(
            0.5, -0.22,
            f"Coef. de correlación de Pearson = {corr:.2f}",
            transform=ax.transAxes,
            ha='center',
            fontsize=12
        )

# Ajusto automáticamente los espacios entre subplots para mejorar la visualización
plt.tight_layout(h_pad=3.0, w_pad=2.0)

# Muestro la figura final con todos los gráficos
plt.show()
```

# 2. Modelado predictivo

Despues de completar el análisis exploratorio de datos y revisar las relaciones entre las características y las variables objetivo, el siguiente paso consiste en desarrollar modelos de aprendizaje automático que permitan predecir el comportamiento de la turbina en términos de **emisiones** y **rendimiento energético**.

En este caso, las variables objetivo del análisis son:

- **CO**: concentración de Monóxido de Carbono
- **NOX**: concentración de Óxidos de Nitrógeno
- **TEY**: Energía generada por la turbina


## Enfoque de modelado predictivo y criterio de optimización

**Teniendo en cuenta que las tres variables objetivo —**CO**, **NOX** y **TEY**— son continuas y que el objetivo inicial es calcular sus valores numéricos**, este problema se aborda como un problema de **regresión**.

Sin embargo, **el propósito del proyecto** no se limita únicamente a predecir emisiones y generación energética de forma aislada. El interés principal **es avanzar hacia una interpretación más aplicada del sistema, de manera que los modelos permitan identificar qué condiciones de operación se asocian con un mejor desempeño global de la turbina**; es decir, con una mayor generación de energía y, al mismo tiempo, con menores concentraciones de contaminantes.

Por esta razón, en este proyecto se plantea una estrategia de modelado en dos niveles.

### 2.1 Modelos de regresión por separado

En primer lugar, se entrenarán modelos independientes para cada variable objetivo (**CO**, **NOX** y **TEY**). Esto se realiza con el fin de analizar por separado el comportamiento de cada variable, comparar el desempeño de los modelos en cada caso y entender con mayor claridad qué tan predecible resulta cada una de ellas a partir de las variables de entrada.

Este enfoque facilita además la interpretación de resultados, ya que no necesariamente las emisiones y la energía generada responden de la misma manera a las variables operativas y ambientales. En otras palabras, modelar cada objetivo por separado permite estudiar con más detalle la sensibilidad de cada variable frente a cambios en presión, temperatura, humedad y demás condiciones de operación.

Para llevar a cabo esta etapa, se utilizará una función que irá variando cada variable objetivo y entrenando los modelos correspondientes de forma iterativa. A su vez, en cada iteración se registrarán en una tabla o **DataFrame de resultados** las principales métricas de evaluación obtenidas —**MAE**, **RMSE** y **R²**— con el propósito de facilitar la comparación entre modelos y variables objetivo.

### 2.2 Modelo de regresión multisalida

Posteriormente, se evaluará un enfoque de **regresión multisalida**, en el cual un mismo modelo intentará predecir simultáneamente las tres variables objetivo. Este enfoque resulta especialmente interesante porque permite estudiar el problema desde una perspectiva más integrada, considerando que las emisiones y el rendimiento energético forman parte del mismo sistema físico y pueden estar relacionados entre sí.

La comparación entre ambos enfoques permitirá evaluar si resulta más conveniente modelar cada variable por separado o si existe valor en un enfoque conjunto capaz de capturar dependencias entre múltiples salidas dentro del sistema. por lo tanto, Los resultados obtenidos con el modelo multisalida también se incorporarán en la misma tabla de resultados construida anteriormente.

## Selección de modelos

Los modelos seleccionados para esta etapa son:

- **Regresión lineal**
- **Random Forest Regressor**
- **XGBoost Regressor**

La **regresión lineal** se utilizará como modelo base, ya que ofrece una referencia simple e interpretable del comportamiento de los datos y permite verificar si existe una relación lineal importante entre las variables predictoras y las variables objetivo.

- **R² (coeficiente de determinación)**, para estimar qué proporción de la variabilidad de la variable objetivo logra explicar el modelo

Estas métricas permitirán comparar de forma objetiva los distintos modelos y seleccionar aquellos que ofrezcan mejores resultados tanto en términos predictivos como en capacidad de interpretación.

## Importo las librerías necesarias para entrenar los modelos mencionados:

```python
# Importo train_test_split para separar los datos en entrenamiento y prueba
from sklearn.model_selection import train_test_split

# Importo las métricas que utilizaré para evaluar los modelos de regresión
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# modelo de regresión lineal como baseline
from sklearn.linear_model import LinearRegression

# Random Forest Regressor 
from sklearn.ensemble import RandomForestRegressor

# XGBoost
from xgboost import XGBRegressor

# MultiOutputRegressor para trabajar con regresión multisalida
from sklearn.multioutput import MultiOutputRegressor
```

## 2.1.1 Separo las características y las variables objetivo:

```python
# Creo una lista con las características que utilizaré para entrenar los modelos
features = ['AT', 'AP', 'AH', 'AFDP', 'GTEP', 'TIT', 'TAT', 'CDP']

# Creo una lista con las variables objetivo continuas que quiero predecir
targets = ['CO', 'NOX', 'TEY']

# Separo el conjunto de características en X
X = df0[features]

# Separo el conjunto de variables objetivo en y
y = df0[targets]
```

## 2.1.3 Separo los datos para la regresión por separado:

```python
# Creo un diccionario vacío para guardar los conjuntos de datos separados por variable objetivo
datasets_por_objetivo = {}

# Recorro cada variable objetivo para crear un split independiente
for target in targets:
    
    # Selecciono la variable objetivo actual
    y_single = df0[target]
    
    # Separo los datos en entrenamiento y prueba para esa variable objetivo
    X_train, X_test, y_train, y_test = train_test_split(
        X, y_single, test_size=0.2, random_state=12345
    )
    
    # Guardo cada conjunto en un diccionario para reutilizarlo más adelante
    datasets_por_objetivo[target] = {
        'X_train': X_train,
        'X_test': X_test,
        'y_train': y_train,
        'y_test': y_test
    }

# Imprimo un ejemplo para verificar que el diccionario se creó correctamente
print(datasets_por_objetivo['CO'].keys())
```

## 2.1.3 Función para entrenar y probar los modelos de regresión por separado:

```python
# Creo una función para entrenar y evaluar los modelos de regresión escogidos
# sobre cada una de las variables objetivo por separado
def evaluar_modelos_regresion(X, y, modelos, test_size=0.2, random_state=12345):
    
    # Creo una lista vacía donde iré guardando los resultados de cada modelo
    resultados = []
    
    # Recorro cada variable objetivo dentro del DataFrame y
    # entreno los modelos uno por uno para esa variable
    for target in y.columns:
        
        # Selecciono la variable objetivo actual
        y_target = y[target]
        
        # Separo los datos en entrenamiento y prueba
        X_train, X_test, y_train, y_test = train_test_split(
            X, y_target, test_size=test_size, random_state=random_state
        )
        
        # Recorro cada modelo definido en el diccionario de modelos
        for nombre_modelo, modelo in modelos.items():
            
            # Entreno el modelo con los datos de entrenamiento
            modelo.fit(X_train, y_train)
            
            # Genero predicciones sobre el conjunto de prueba
            y_pred = modelo.predict(X_test)
            
            # Calculo el MAE para medir el error promedio absoluto
            mae = mean_absolute_error(y_test, y_pred)
            
            # Calculo el RMSE para penalizar más los errores grandes
            rmse = np.sqrt(mean_squared_error(y_test, y_pred))
            
            # Calculo el R² para estimar qué proporción de la variabilidad explica el modelo
            r2 = r2_score(y_test, y_pred)
            
            # Guardo los resultados del modelo actual en forma de diccionario
            resultados.append({
                'tipo_modelo': 'regresion_por_separado',
                'variable_objetivo': target,
                'modelo': nombre_modelo,
                'MAE': mae,
                'RMSE': rmse,
                'R2': r2
            })
    
    # Convierto la lista de resultados en un DataFrame para facilitar su comparación
    resultados_df = pd.DataFrame(resultados)
    
    # Devuelvo el DataFrame final con todas las métricas obtenidas
    return resultados_df
```

## Teniendo en cuenta que los modelos que dieron los mejores resultados fueron XBRegressor y Random Forest, voy a enfocarme en estos y compararlos variando sus hiperparametros 

```python
# Importo ParameterGrid para recorrer las combinaciones de hiperparámetros
from sklearn.model_selection import ParameterGrid

# Defino los hiperparámetros para Random Forest
param_grid_rf = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 5, 10, 15]
}

# Defino losv hiperparámetros para XGBoost
param_dist_xgb = {
    'n_estimators': [100, 200, 300],
    'learning_rate': [0.01, 0.05, 0.1],
    'max_depth': [5, 10, 15]
}
```

## Nueva función para evaluar los mejores 2 modelos variando hiperparametros:

```python
# Creo una función para entrenar y evaluar Random Forest y XGBoost
# probando distintas combinaciones de hiperparámetros para cada variable objetivo
def evaluar_modelos_con_hiperparametros(X, y, param_grid_rf, param_dist_xgb, test_size=0.2, random_state=42):
    
    # Creo una lista vacía donde iré guardando los resultados
    resultados = []
    
    # Recorro cada variable objetivo del DataFrame y
    # realizo un entrenamiento independiente para cada una
    for target in y.columns:
        
        # Selecciono la variable objetivo actual
        y_target = y[target]
        
        # Separo los datos en entrenamiento y prueba
        X_train, X_test, y_train, y_test = train_test_split(
            X, y_target, test_size=test_size, random_state=random_state
        )
        
        # Recorro todas las combinaciones posibles de hiperparámetros para Random Forest
        for params_rf in ParameterGrid(param_grid_rf):
            
            # Configuro el modelo Random Forest con la combinación actual
            modelo_rf = RandomForestRegressor(
                random_state=random_state,
                **params_rf
            )
            
            # Entreno el modelo con los datos de entrenamiento
            modelo_rf.fit(X_train, y_train)
            
            # Genero predicciones sobre el conjunto de prueba
            y_pred_rf = modelo_rf.predict(X_test)
            
            # Calculo el MAE para medir el error promedio absoluto
            mae_rf = mean_absolute_error(y_test, y_pred_rf)
            
            # Calculo el RMSE para penalizar más los errores grandes
            rmse_rf = np.sqrt(mean_squared_error(y_test, y_pred_rf))
            
            # Calculo el R² para estimar la variabilidad explicada por el modelo
            r2_rf = r2_score(y_test, y_pred_rf)
            
            # Guardo los resultados de esta combinación de Random Forest
            resultados.append({
                'tipo_modelo': 'regresion_por_separado',
                'variable_objetivo': target,
                'modelo': 'RandomForestRegressor',
                'hiperparametros': params_rf,
                'MAE': mae_rf,
                'RMSE': rmse_rf,
                'R2': r2_rf
            })
        
        # Recorro todas las combinaciones posibles de hiperparámetros para XGBoost
        for params_xgb in ParameterGrid(param_dist_xgb):
            
            # Configuro el modelo XGBoost con la combinación actual
            modelo_xgb = XGBRegressor(
                random_state=random_state,
                objective='reg:squarederror',
                **params_xgb
            )
            
            # Entreno el modelo con los datos de entrenamiento
            modelo_xgb.fit(X_train, y_train)
            
            # Genero predicciones sobre el conjunto de prueba
            y_pred_xgb = modelo_xgb.predict(X_test)
            
            # Calculo el MAE para medir el error promedio absoluto
            mae_xgb = mean_absolute_error(y_test, y_pred_xgb)
            
            # Calculo el RMSE para penalizar más los errores grandes
            rmse_xgb = np.sqrt(mean_squared_error(y_test, y_pred_xgb))
            
            # Calculo el R² para estimar la variabilidad explicada por el modelo
            r2_xgb = r2_score(y_test, y_pred_xgb)
            
            # Guardo los resultados de esta combinación de XGBoost
            resultados.append({
                'tipo_modelo': 'regresion_por_separado',
                'variable_objetivo': target,
                'modelo': 'XGBRegressor',
                'hiperparametros': params_xgb,
                'MAE': mae_xgb,
                'RMSE': rmse_xgb,
                'R2': r2_xgb
            })
    
    # Convierto la lista de resultados en un DataFrame
    resultados_df = pd.DataFrame(resultados)
    
    # Devuelvo el DataFrame final para facilitar su análisis y comparación
    return resultados_df
```

## Pruebo la función 2:

```python
# Ejecuto la función para evaluar los modelos con sus distintas combinaciones
resultados_modelos = evaluar_modelos_con_hiperparametros(X, y, param_grid_rf, param_dist_xgb)

# Muestro los resultados
resultados_modelos
```

Una vez evaluado el desempeño predictivo de los modelos, el siguiente paso consiste en trasladar ese análisis hacia una perspectiva más aplicada. En lugar de limitarse únicamente a predecir emisiones y generación energética, el objetivo ahora es identificar las condiciones operativas y ambientales bajo las cuales la turbina presenta un comportamiento más favorable.

Para ello, se construirá un criterio de clasificación basado en la propia distribución del dataset. En este caso, se considerarán como condiciones favorables aquellas observaciones donde:

- la **energía generada por la turbina (TEY)** se ubique en el cuartil superior de su distribución,
- las emisiones de **CO** se ubiquen en el cuartil inferior,
- y las emisiones de **NOX** también se ubiquen en el cuartil inferior.

A partir de este criterio se creará una variable denominada **condición óptima**, la cual permitirá aislar y analizar las observaciones que combinan simultáneamente un alto desempeño energético y una menor carga contaminante.

De esta manera, **la optimización dentro del proyecto** no **se entiende** como un control directo en tiempo real sobre la turbina, sino **como la identificación de patrones operativos que puedan interpretarse como condiciones más eficientes y ambientalmente más favorables.**


## 3.1 Separo df0 usando los Percentiles y Creo columnas nuevas que convierten las variables objetivo de continuas a categóricas (TRUE/FALSE o 1/0):

```python
# Calculo el percentil 75 de TEY para identificar condiciones de alta generación de energía
tey_q75 = df0['TEY'].quantile(0.75)

# Calculo el percentil 25 de CO para identificar condiciones de baja emisión de monóxido de carbono
co_q25 = df0['CO'].quantile(0.25)

# Calculo el percentil 25 de NOX para identificar condiciones de baja emisión de óxidos de nitrógeno
nox_q25 = df0['NOX'].quantile(0.25)

# Imprimo los umbrales utilizados para clasificar las observaciones
print(f"Percentil 75 de TEY: {tey_q75:.4f}")
print(f"Percentil 25 de CO: {co_q25:.4f}")
print(f"Percentil 25 de NOX: {nox_q25:.4f}")


# Creo una columna categórica para identificar observaciones con alta generación de energía
df0['TEY_alto'] = np.where(df0['TEY'] >= tey_q75, 1, 0)

# Creo una columna categórica para identificar observaciones con bajas emisiones de CO
df0['CO_bajo'] = np.where(df0['CO'] <= co_q25, 1, 0)

# Creo una columna categórica para identificar observaciones con bajas emisiones de NOX
df0['NOX_bajo'] = np.where(df0['NOX'] <= nox_q25, 1, 0)

# Creo una columna final para identificar condiciones simultáneamente favorables
df0['condicion_optima'] = np.where(
    (df0['TEY_alto'] == 1) &
    (df0['CO_bajo'] == 1) &
    (df0['NOX_bajo'] == 1),
    1,
    0
)

# Reviso cuántas observaciones fueron clasificadas como óptimas
print("\nDistribución de la variable condicion_optima:")
print(df0['condicion_optima'].value_counts())
```

## 3.2 Creo un nuevo DataFrame que contiene únicamente las observaciones consideradas óptimas

```python
# Creo un nuevo DataFrame df_optimo
df_optimo = df0[df0['condicion_optima'] == 1].copy()

# Imprimo la cantidad de registros óptimos encontrados
print(f"\nNúmero de observaciones clasificadas como óptimas: {df_optimo.shape[0]}")

# Muestro las primeras filas del subconjunto óptimo
df_optimo.head()
```
